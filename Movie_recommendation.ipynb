##**Column** **filtering**

```python
def convert(obj): #The convert function takes a string as input and attempts to
  try:            #convert it into a list of names extracted from a list of dictionaries.
    l=[]
    for i in ast.literal_eval(obj): #ast.literal_eval(obj): This attempts to evaluate the input string obj
        l.append(i["name"])         #as a Python expression and returns the resulting value.
    return l
  except (SyntaxError, ValueError):
    return []
```

```python
movies_df["genres"]=movies_df["genres"].apply(convert)
movies_df["keywords"]=movies_df["keywords"].apply(convert)
```

```python
movies_df.head(2)
```

```python
def convert1(obj):
  try:
    l = []
    count = 0
    for i in ast.literal_eval(obj):
      if count != 3:
        l.append(i["name"])
      count += 1
    return l
  except SyntaxError:
    return []
```

```python
movies_df["cast"]=movies_df["cast"].apply(convert1)
```

```python
movies_df.head(2)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
      if i["job"]=="Director":
        l.append(i["name"])
    return l
```

```python
movies_df["crew"]=movies_df["crew"].apply(fetch_director)
```

```python
movies_df.head(2)
```

##**Feature Extraction**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')

 #used to transform a given text into a vector on the basis of freqency count
```

```python
cv.fit_transform(new_df['target']).toarray().shape
```

**cv.fit_transform:** This method of the CountVectorizer object takes a dataset as input and performs two operations:


**Fit:** It learns the vocabulary of the dataset, which in this case is the unique words present in the target column of the new_df DataFrame.


**Transform:** It transforms the dataset into a matrix of token counts, where each row represents a document (in this case, a row in the target column) and each column represents a unique word from the vocabulary.

```python
vectors=cv.fit_transform(new_df['target']).toarray()
```

```python
vectors[0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

#The PorterStemmer class is used for stemming words, which is the process of reducing
#words to their root form. For example, the word "running" would be stemmed to "run".
```

```python
def stem(text):
  l=[]
  for i in text.split():
    l.append(ps.stem(i))
  return l
```

```python
new_df['target']=new_df['target'].apply(stem)
```

```python
#!pip install scikit-learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity


#cosine_similarity measures the similarity between two vectors of an inner product space.
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
# Find the five most similar documents to the first document

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#This code snippet essentially recommends five movies that are most similar to the provided movie based on their cosine similarities. This can be useful

def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
